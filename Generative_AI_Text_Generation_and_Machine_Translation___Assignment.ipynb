#Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

- Probabilistic modeling is used in generative models to learn the probability distribution of the data.
The model tries to understand how the data is generated and then uses that probability distribution to create new samples similar to the original data.

- By learning this distribution, the model can generate new realistic data points, not just classify existing ones.

- Examples of generative models include Generative Adversarial Network, Variational Autoencoder, and GPT (Generative Pre-trained Transformer).

**Difference Between Generative and Discriminative Models**

   - Generative models learn the joint probability P(X,Y), while discriminative models learn the conditional probability P(Y∣X).

   - Generative models generate new data, while discriminative models predict labels or classify data.

   - Generative models focus on how the data is created, whereas discriminative models focus on separating different classes.

   - Generative models are useful for data generation, simulation, and unsupervised learning.

   - Discriminative models are mainly used for classification and prediction tasks.
   - Examples of discriminative models include Logistic Regression and Support Vector Machine.

#Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

- **Difference Between Autoencoders and Variational Autoencoders in Text Generation**

1. Basic Concept

    - Autoencoders: Neural networks that compress input text into a latent representation and then reconstruct the same text.

    - Variational Autoencoders: A probabilistic version of autoencoders that learn a distribution of the latent space instead of a single fixed representation.

2. Latent Representation

    - Autoencoders: Learn a deterministic latent vector.

    - VAEs: Learn a probability distribution (mean and variance) of latent variables.

3. Data Generation Ability

    - Autoencoders: Mainly used for reconstruction, not good at generating new text.

    - VAEs: Can generate new text samples by sampling from the learned latent distribution.

4. Handling Variability

    - Autoencoders: Do not model uncertainty in data.

    - VAEs: Use probabilistic modeling to capture uncertainty and variability in text data.

5. Training Approach

    - Autoencoders: Minimize reconstruction loss.

    - VAEs: Optimize reconstruction loss + KL divergence to ensure a structured latent space.

6. Examples

    - Autoencoder: Autoencoder

    - VAE: Variational Autoencoder

#Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?
1. Attention mechanism helps the model focus on important words in the input sentence while translating.

2. In Neural Machine Translation, an encoder first converts the source sentence into hidden representations, and a decoder generates the translated sentence.

3. The attention mechanism calculates attention weights for each word in the input sentence. These weights show which words are most relevant for generating the current output word.

4. The model then creates a context vector, which is a weighted combination of encoder outputs based on the attention scores.

5. This context vector helps the decoder focus on the correct part of the source sentence when generating each word in the translation.

6. Attention is widely used in modern models like Transformer and BERT.

- **Attention Mechanisms are Critical because** :-

1. They help the model focus on relevant words during translation.

2. They solve the long sentence problem by not forcing the model to rely on a single fixed context vector.

3. They improve translation accuracy and context understanding.

4. They allow models to capture relationships between words in different positions.

5. They form the foundation of modern architectures like the Transformer, which power many advanced language models.

#Question 6: Use the following small text dataset to train a simple Variational Autoencoder (VAE) for text reconstruction:
**["The sky is blue", "The sun is bright", "The grass is green",
"The night is dark", "The stars are shining"]**

*1. Preprocess the data (tokenize and pad the sequences).

2. Build a basic VAE model for text reconstruction.

3. Train the model and show how it reconstructs or generates similar sentences.
Include your code, explanation, and sample outputs*

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input, Dense, Lambda, Embedding, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K

texts = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

vocab_size = len(tokenizer.word_index) + 1

embedding_dim = 8
latent_dim = 2

inputs = Input(shape=(max_len,))
embedding = Embedding(vocab_size, embedding_dim)(inputs)
flat = Flatten()(embedding)

z_mean = Dense(latent_dim)(flat)
z_log_var = Dense(latent_dim)(flat)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon

z = Lambda(sampling)([z_mean, z_log_var])

decoder = Dense(max_len, activation="softmax")
outputs = decoder(z)

vae = Model(inputs, outputs)

reconstruction_loss = tf.keras.losses.sparse_categorical_crossentropy(inputs, outputs)
kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)

vae_loss = K.mean(reconstruction_loss + kl_loss)

vae.add_loss(vae_loss)
vae.compile(optimizer="adam")

vae.fit(
    padded_sequences,
    padded_sequences,
    epochs=200,
    batch_size=2
)

predictions = vae.predict(padded_sequences)
```

#Question 8: Implement a simple attention-based encoder-decoder model for English-to-Spanish translation using Tensorflow or PyTorch.


```python
import tensorflow as tf
from tensorflow.keras.layers import Input, LSTM, Dense, Embedding, Attention
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# ---------------------------
# 1. Small English-Spanish Dataset
# ---------------------------
english_texts = [
    "Hello",
    "How are you",
    "I am fine",
    "Thank you",
    "Good morning"
]

spanish_texts = [
    "Hola",
    "Cómo estás",
    "Estoy bien",
    "Gracias",
    "Buenos días"
]

# ---------------------------
# 2. Tokenization
# ---------------------------
eng_tokenizer = Tokenizer(filters='')
eng_tokenizer.fit_on_texts(english_texts)
eng_sequences = eng_tokenizer.texts_to_sequences(english_texts)
eng_maxlen = max(len(seq) for seq in eng_sequences)
eng_vocab_size = len(eng_tokenizer.word_index) + 1
eng_padded = pad_sequences(eng_sequences, maxlen=eng_maxlen, padding='post')

spa_tokenizer = Tokenizer(filters='')
spa_tokenizer.fit_on_texts(spanish_texts)
spa_sequences = spa_tokenizer.texts_to_sequences(spanish_texts)
spa_maxlen = max(len(seq) for seq in spa_sequences)
spa_vocab_size = len(spa_tokenizer.word_index) + 1
spa_padded = pad_sequences(spa_sequences, maxlen=spa_maxlen, padding='post')

# ---------------------------
# 3. Model Parameters
# ---------------------------
embedding_dim = 16
latent_dim = 32

# ---------------------------
# 4. Encoder
# ---------------------------
encoder_inputs = Input(shape=(eng_maxlen,))
enc_emb = Embedding(eng_vocab_size, embedding_dim)(encoder_inputs)
encoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(enc_emb)

# ---------------------------
# 5. Decoder
# ---------------------------
decoder_inputs = Input(shape=(spa_maxlen,))
dec_emb = Embedding(spa_vocab_size, embedding_dim)(decoder_inputs)
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=[state_h, state_c])

# ---------------------------
# 6. Attention
# ---------------------------
attention_layer = Attention()
attention_output = attention_layer([decoder_outputs, encoder_outputs])

# Combine attention output and decoder LSTM output
decoder_concat = tf.keras.layers.Concatenate(axis=-1)([decoder_outputs, attention_output])

# Final Dense layer for predictions
decoder_dense = Dense(spa_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_concat)

# ---------------------------
# 7. Define and Compile Model
# ---------------------------
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

# ---------------------------
# 8. Prepare Targets (shifted)
# ---------------------------
# Decoder targets need an extra dimension
decoder_targets = np.expand_dims(spa_padded, -1)

# ---------------------------
# 9. Train Model
# ---------------------------
model.fit([eng_padded, spa_padded], decoder_targets, epochs=300, batch_size=2)

# ---------------------------
# 10. Prediction Function
# ---------------------------
def translate(sentence):
    seq = eng_tokenizer.texts_to_sequences([sentence])
    seq = pad_sequences(seq, maxlen=eng_maxlen, padding='post')
    dec_input = np.zeros((1, spa_maxlen))
    output_tokens = model.predict([seq, dec_input])
    word_indices = np.argmax(output_tokens[0], axis=-1)
    words = [word for idx in word_indices if idx != 0 for word, i in spa_tokenizer.word_index.items() if i == idx]
    return ' '.join(words)

# Example
print(translate("Hello"))
```

#Question 9: Use the following short poetry dataset to simulate poem generation with apre-trained GPT model:
- ["Roses are red, violets are blue,",

- Using this dataset as a reference for poetic structure and language, generate a new 2-4 line poem using a pre-trained GPT model (such as GPT-2).
- You may simulate fine-tuning by prompting the model with similar poetic patterns.
Include your code, the prompt used, and the generated poem in your answer.


```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Small poetry dataset (used for inspiration in the prompt)
poems = [
    "Roses are red, violets are blue,",
    "Sugar is sweet, and so are you.",
    "The moon glows bright in silent skies,",
    "A bird sings where the soft wind sighs."
]

# Prompt to simulate fine-tuning / poetic style
prompt = (
    "Using the style of the following poems:\n"
    "'Roses are red, violets are blue,'\n"
    "'Sugar is sweet, and so are you.'\n"
    "'The moon glows bright in silent skies,'\n"
    "'A bird sings where the soft wind sighs.'\n\n"
    "Generate a new 2-4 line poem:"
)

# Encode prompt
inputs = tokenizer.encode(prompt, return_tensors="pt")

# Generate text
outputs = model.generate(
    inputs,
    max_length=100,
    num_return_sequences=1,
    no_repeat_ngram_size=2,
    temperature=0.8
)

# Decode generated text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Generated Poem:\n")
print(generated_text.split("Generate a new 2-4 line poem:")[1].strip())
```
