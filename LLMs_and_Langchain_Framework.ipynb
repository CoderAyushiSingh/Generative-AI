##Question 1: What are Large Language Models (LLMs) and how do they function?


- Large Language Models (LLMs) are advanced AI models designed to understand and generate human-like text. They are trained on very large datasets of books, websites, articles, and other text to learn language patterns.

Function:-

1. Training on large text data
    -  The model learns grammar, context, and relationships between words.

2. Tokenization
    -  Text is broken into smaller units called tokens.

3. Transformer architecture
    -  LLMs use transformers with attention mechanisms to understand context in sentences.

4. Next-word prediction
   -  The model predicts the most probable next word/token based on previous words.

5. Generation
    -  By repeatedly predicting the next token, the model generates full sentences or paragraphs.

- Examples: GPT-4, BERT, LLaMA.

- LLMs work by learning patterns from massive text data and predicting the next word in a sequence to generate meaningful language.

##Question 5: Compare and contrast Langchain and LamaIndex. What unique problems does each solve?


- LangChain vs LlamaIndex

    - Both are frameworks used to build applications with Large Language Models, but they focus on different problems.

**LangChain**
- LangChain is mainly used to build complete LLM-powered applications. It provides tools to manage prompts, connect models with APIs, maintain conversation memory, and create multi-step workflows.

- **Problems it solves:**

    - Managing prompt templates and chains

    - Building AI agents that can use tools

    - Handling conversation memory

    - Integrating LLMs with external services like databases and APIs

- Example:
    -  A customer support chatbot that can search a database, call an API, and generate responses.

**LlamaIndex**

- LlamaIndex focuses on connecting LLMs with large external datasets. It organizes documents into indexes so that the model can retrieve relevant information before generating an answer.

- **Problems it solves:**

    - Efficient document indexing and retrieval

    - Working with large document collections

    - Building Retrieval-Augmented Generation (RAG) systems

    - Connecting LLMs with PDFs, databases, and knowledge bases

- Example:
    - A system where users upload company documents and ask questions, and the AI answers using those documents.

 ##Question 6: Implement a basic Langchain pipeline using OpenAI’s LLM to answer questions based on a user input prompt.
   

```python
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Initialize the LLM
llm = OpenAI(openai_api_key="YOUR_API_KEY")

# Create a prompt template
template = """
You are a helpful assistant.
Answer the following question clearly.

Question: {question}
Answer:
"""

prompt = PromptTemplate(
    input_variables=["question"],
    template=template
)

# Create LangChain pipeline
chain = LLMChain(
    llm=llm,
    prompt=prompt
)

# User input
user_question = input("Enter your question: ")

# Generate response
response = chain.run(user_question)

print("Answer:", response)
```

##Question 7: Integrate Langchain with a third-party API (e.g., weather, news) and show how responses can be generated via LLMs.



```python
import requests
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Initialize OpenAI LLM
llm = OpenAI(openai_api_key="YOUR_API_KEY")

# Function to call third-party Weather API
def get_weather(city):
    api_key = "YOUR_WEATHER_API_KEY"
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url).json()

    temp = response["main"]["temp"]
    description = response["weather"][0]["description"]

    return f"The temperature in {city} is {temp}°C with {description}."

# Prompt template
template = """
You are a helpful assistant.

Weather Data: {weather_info}

Explain the weather condition to the user in a simple way.
"""

prompt = PromptTemplate(
    input_variables=["weather_info"],
    template=template
)

# LangChain pipeline
chain = LLMChain(llm=llm, prompt=prompt)

# User input
city = input("Enter city name: ")

weather_info = get_weather(city)

# Generate LLM response
response = chain.run(weather_info)

print(response)
```

##Question 8: Create a LamaIndex implementation that indexes a local text file and retrieves answers from it.

```python
from llama_index import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms import OpenAI

# Load documents from a local folder
documents = SimpleDirectoryReader("data").load_data()

# Create an index
index = VectorStoreIndex.from_documents(documents)

# Create query engine
query_engine = index.as_query_engine()

# Ask question based on the indexed file
response = query_engine.query("What is the main topic of the document?")

print(response)
```

##Question 9: Demonstrate combining Langchain with LamaIndex to create a simple document-based Q&A chatbot.

```python
from llama_index import VectorStoreIndex, SimpleDirectoryReader
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import FAISS

# Step 1: Load local documents
documents = SimpleDirectoryReader("data").load_data()

# Step 2: Create LlamaIndex
index = VectorStoreIndex.from_documents(documents)

# Convert index to retriever
retriever = index.as_retriever()

# Step 3: Initialize LLM
llm = OpenAI(openai_api_key="YOUR_API_KEY")

# Step 4: Create LangChain QA pipeline
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    chain_type="stuff"
)

# Step 5: Chat loop
while True:
    question = input("Ask a question (type exit to stop): ")

    if question.lower() == "exit":
        break

    answer = qa_chain.run(question)
    print("Bot:", answer)
```

##Question 10: A legal firm wants to use AI to summarize large volumes of legal documents and retrieve relevant information quickly. Propose a solution using Langchain and LamaIndex, and explain how it would work in practice.


- Document Summarization & Retrieval System

**Architecture**

1. Document Ingestion

   - All legal documents (contracts, case files, agreements) are stored in a folder or cloud storage.

   - LlamaIndex reads and parses these documents.

2. Indexing

   - Use LlamaIndex to convert documents into embeddings and store them in a vector database (like FAISS or Pinecone).

   - Each document chunk is indexed for fast retrieval.

3.  LLM Integration via LangChain

    - LangChain acts as the application layer.

    - It connects the retriever (LlamaIndex) with the LLM to create a Retrieval-Augmented Generation (RAG) pipeline.

    - The LLM generates summaries, answers, or highlights relevant clauses from retrieved documents.

4. User Interface

    - Legal staff can enter queries in natural language, e.g.:

        - “Summarize all clauses related to termination.”

        - “Find precedents for contract disputes in 2023.”

   - The system retrieves relevant chunks and produces concise, human-readable answers.

   



```python
from llama_index import VectorStoreIndex, SimpleDirectoryReader
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA

documents = SimpleDirectoryReader("legal_docs").load_data()

index = VectorStoreIndex.from_documents(documents)
retriever = index.as_retriever()

llm = OpenAI(openai_api_key="YOUR_API_KEY")

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    chain_type="stuff"
)

query = "Summarize clauses about dispute resolution."
answer = qa_chain.run(query)
print(answer)
```
